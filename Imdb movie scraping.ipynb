```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
r=requests.get("https://www.imdb.com/chart/top/")
```

```python
r
```

```python
r.text
```

```python
soup=BeautifulSoup(r.text,"html.parser")
```

```python
movies=soup.find("tbody",class_="lister-list").find_all("tr")
```

```python
movies
```

```python
movie_data=[]
for i in movies:
    movie_name=i.find("td",class_="titleColumn").a.text
    movie_rating=i.find("td",class_="ratingColumn").strong.text
    movie_rank=i.find("td",class_="titleColumn").get_text(strip=True)[0]
    movie_year=i.find("td",class_="titleColumn").span.text[1:-1]
    movie_data.append((movie_rank,movie_name,movie_year,movie_rating))
```

```python
movie_data
```

```python
import pandas as pd 
```

```python
df=pd.DataFrame(data=movie_data,columns=["Rank","Movie_Name","Year","Rating"]) 
```

```python
df.to_csv("movie_details.csv")
```
